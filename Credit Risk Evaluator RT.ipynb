### Step 1: Load the Datasets

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

### Step 2: Preprocessing Data

For each set of data convert categorical data to numeric and separate target feature for training data
THEN convert categorical data to numeric and separate target feature for testing data

#### Training Data

```python
X_dummiestrain = pd.get_dummies(train_df)
print(X_dummiestrain.columns)
X_dummiestrain
```

```python

X_dummiestrain.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
X_dummiestrain
```

#### Testing Data

```python
X_dummiestest = pd.get_dummies(test_df)
print(X_dummiestest.columns)

X_dummiestest
```

```python
X_dummiestest.drop('Unnamed: 0', axis=1, inplace=True)
```

#### add missing dummy variables to testing set to make even amount of columns

```python
X_dummiestest['debt_settlement_flag_Y'] = 0
```

```python
X_dummiestest
```

### Step 3: separate target feature for testing data and create  X_train, X_test, y_train, y_test

```python
X_train = X_dummiestrain.drop('loan_status_low_risk', axis=1)
X_train
```

```python
y_train = LabelEncoder().fit_transform(X_dummiestrain['loan_status_low_risk'])
y_train
```

```python
X_test = X_dummiestest.drop('loan_status_low_risk', axis=1)
X_test
```

```python
y_test = LabelEncoder().fit_transform(X_dummiestest['loan_status_low_risk'])
y_test
```

### Step 4: Train the Logistic Regression model on the unscaled data and print the model score

```python
classifier = LogisticRegression()
classifier
```

```python
 classifier.fit(X_train, y_train)
```

```python
 print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

### Step 5: Train a Random Forest Classifier model and print the model score

```python
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train, y_train)
print(f'Training Score: {clf.score(X_train, y_train)}')
print(f'Testing Score: {clf.score(X_test, y_test)}')
```

### Step 6: Scale the data

```python
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

### Step 7: Train the Logistic Regression model on the scaled data and print the model score

```python
 classifier.fit(X_train_scaled, y_train)
```

```python
 print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

### Step 8: Train a Random Forest Classifier model on the scaled data and print the model score

```python
clf = RandomForestClassifier(random_state=1, n_estimators=500).fit(X_train_scaled, y_train)
print(f'Training Score: {clf.score(X_train_scaled, y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, y_test)}')
```

### Conclusion

It seems as though the Random Forest Classifier works best but there also may be an error as it scored 100 both times in the test and training score. The Logistic Regression model became much more accurate after the data was scaled
